**MODEL**

```python
from sklearn.model_selection import train_test_split  
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
X = dataset.drop(columns='species')
y = dataset['species']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

Tes dataset yang digunakan sebesar 20 %

**K-Nearest Neighbors**

```python
from sklearn.neighbors import KNeighborsClassifier
i_range= list(range(1,10)) 
scores = []

for i in i_range:
    model_knn=KNeighborsClassifier(n_neighbors=i) 
    model_knn.fit(X_train, y_train)  
    y_predict=model_knn.predict(X_test) 
    scores.append(accuracy_score(y_test, y_predict))
```

```python
plt.plot(i_range, scores)
plt.xlabel('Value i')
plt.ylabel('Accuracy score i')
plt.title('Accuracy score i in KNN')
plt.tight_layout
plt.show()
```

```python
model_knn=KNeighborsClassifier(n_neighbors=8)
model_knn.fit(X_train, y_train)  
y_predict=model_knn.predict(X_test)
print(classification_report(y_test, y_predict))

from sklearn.metrics import accuracy_score
from __future__ import division
print('Akurasi KNN: ' + "{:.1%}".format(accuracy_score(y_predict, y_test)))
```

Hasil akurasi menggunakan metode KNN diperoleh sebesar 100 %

**Naive Bayes**

```python
from sklearn.naive_bayes import GaussianNB

naive_bayes_model = GaussianNB()
naive_bayes_model.fit(X_train, y_train)
```

```python
y_predict = naive_bayes_model.predict(X_test)
```

```python
print(classification_report(y_test, y_predict))
print(confusion_matrix(y_test, y_predict))

print('Akurasi Naive Bayes: ' + "{:.1%}".format(accuracy_score(y_predict, y_test)))
```

Hasil akurasi menggunakan metode Naive Bayes diperoleh sebesar 96.7%
